# we can fill null value by mean median and mode
again load data

```python
df1=pd.read_csv('empl.csv')
df1
```

```python
df1.isnull().sum()
```

```python
# we have one more method to remove null values
# we dont need to remove null value we can fill with mean median and mode 
df1['Age']=df1['Age'].fillna((df1['Age'].median()))
```

```python
df1.isnull().sum()
```

```python
df1['Salary']=df1['Salary'].fillna((df1['Salary'].mean()))
```

```python
df1.isnull().sum()
```

```python
df1['City']=df1['City'].fillna((df1['City'].mode()[0]))
#its mode()[0] means take first top most city name with high frequency
```

```python
df1.isnull().sum()
```

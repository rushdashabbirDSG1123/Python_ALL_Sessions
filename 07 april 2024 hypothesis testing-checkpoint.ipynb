# categorical variable test of independence... chi square test

```python
# checkthe chi square test
import numpy as np
import pandas as pd
import scipy.stats as stats
pd.read_csv('role_website.csv',sep=',')
```

```python
ds=pd.read_csv('role_website.csv')
ds
```

```python
df3=pd.DataFrame(ds)
df3
```

```python
df3.describe()
```

```python
df3.info()
```

```python
df3.dtypes
```

```python
# create contigency table
#it ise to output in table formate
df_crosstab=pd.crosstab(df3['role'],df3['website'],margins=True)
print(df_crosstab)
```

```python
df_crosstab.columns=['facebook','linkedin','row_totals']
df_crosstab.index=['student','teacher','col_totals']
df_crosstab
```

```python
#obserbed table
observed=df_crosstab.iloc[0:2,0:2]
observed
```

```python
#expected table
#E= raw total* column total/ grand total
#E=13*16/34
#E1=13*18/34
expected=np.outer(df_crosstab["row_totals"][0:2],df_crosstab.loc["col_totals"][0:2])/34
expected
```

```python
from scipy.stats import chi2_contingency
```

```python
chi2,p,dof,expected=stats.chi2_contingency(observed=observed)

print('chi square test =',chi2)
print('P-value =',p)
print('dof =',dof)
print('expected =',expected)
```

```python
# we can write like this also
from scipy.stats import chi2_contingency
observed=[[10,20,30],[6,9,17]]
chi2, p, dof, expected = chi2_contingency(observed)

print('chi square test =',chi2)
print('P-value =',p)
print('dof =',dof)
print('expected =',expected)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error
from sklearn.model_selection import train_test_split

from sklearn.linear_model import SGDRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import DecisionTreeClassifier
```

```python
x2_train,x2_test,y2_train,y2_test=train_test_split(x2,y2,test_size=.20,random_state=48)
```

```python
x2_train.shape
```

```python
y2_train.shape
```

```python
x2_test.shape
```

```python
x2_test.shape
```

```python
lr=LinearRegression()
lr.fit(x2_train,y2_train)
```

```python
# again write code in 1 time
lr.intercept_
lr.score(x2_train,y2_train)
pred=lr.predict(x2_test)
#now conver pred data into data frame
preddf=pd.DataFrame(data=pred)
preddf
lr.score(x2_train,y2_train)
print('mean_absolute_error :',mean_absolute_error(y2_test,pred))
print('mean_squared_error :',mean_squared_error(y2_test,pred))
print('root_mean_squared_error :',np.sqrt(mean_squared_error(y2_test,pred)))
print("r2 score:",r2_score(y2_test,pred))
```

```python
dtc=DecisionTreeRegressor()
dtc.fit(x2_train,y2_train)
dtc.score(x2_train,y2_train)
preddtc=dtc.predict(x2_test)
print('mean_absolute_error :',mean_absolute_error(y2_test,preddtc))
print('mean_squared_error :',mean_squared_error(y2_test,preddtc))
print('root_mean_squared_error :',np.sqrt(mean_squared_error(y2_test,preddtc)))
print("r2 score:",r2_score(y2_test,preddtc))
```

# Stochastic  Gradient Descent Algorithm

```python
from sklearn.linear_model import SGDRegressor
#Stochastic Gradient Descent Algorithm its use for optimise error
sgd=SGDRegressor()
sgd.fit(x2_train,y2_train)
sgd.score(x2_train,y2_train)
predsgd=sgd.predict(x2_test)
print('mean_absolute_error :',mean_absolute_error(y2_test,predsgd))
print('mean_squared_error :',mean_squared_error(y2_test,predsgd))
print('root_mean_squared_error :',np.sqrt(mean_squared_error(y2_test,predsgd)))
print("r2 score:",r2_score(y2_test,predsgd))
```

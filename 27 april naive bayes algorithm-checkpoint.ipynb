# Here write function for mnb

```python
def predict_mnb(t):
    t=t.reshape(1,-1)
    p=mnb.predict(t)
    print(p)
    if p==0:
        print('setosa')
    elif p==1:
        print('Versicular')
    elif p==2:
        print('virginica')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
from sklearn import preprocessing
#for Unsupervised Machine Learning
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN
import warnings
warnings.filterwarnings("ignore")
from sklearn.cluster import KMeans
from sklearn import preprocessing
```

```python
df=pd.read_csv('Mall_customers.csv')
df.head(10)
```

```python
x=df.iloc[:,3:]
x.tail(8)
```

```python
#here use elbow method
wcss=[]
for i in range (1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++')
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss,marker='x')
plt.title('The Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#filtewr kmean to the datasets
kmeans=KMeans(n_clusters=5,init='k-means++')
y_kmeans=kmeans.fit_predict(x)
print(y_kmeans)
#its return 5 groups
```

# Pipe line

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
import warnings
warnings.filterwarnings("ignore")
```

```python
dfp=pd.read_csv('pima_indian_diabetes.csv')
dfp
```

```python
dfp.keys
```

```python
dfp.columns
```

```python
array=dfp.values
array
```

```python
x=array[:,0:8]
x
```

```python
y=array[:,8]
y
```

```python
#create PipeLine
estimators=[]
estimators.append(("standardize",StandardScaler()))
estimators.append(('dtc',DecisionTreeClassifier()))
model=Pipeline(estimators)
model

#evalute pipeline
result=cross_val_score(model,x,y,cv=5)
print(result.mean())
```

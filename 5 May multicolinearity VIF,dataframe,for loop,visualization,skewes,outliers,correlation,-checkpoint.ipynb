# Multicollinearity

multicolinearity is the type data which we have multiple same type of columns two or more colums having similar values to predict y label so in this case y in notcapable to known which similar columns contioburte most to prediction so this is multicolinearity 
so we do not need multicolinearity in data set we have to need remove that collinearity from the data set with the help of VIF (variance Inflation factor )means we need to detect how much collinearity inside the data set that we can detect that throgh VIF.
VIF is measure the colinearity among predictor variables within a multiple regression it is a measure for the increase of the variance of the parameters estimaters and its also find highl;y collerated variable.

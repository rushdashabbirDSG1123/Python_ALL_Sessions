```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
from sklearn import preprocessing
#for Unsupervised Machine Learning
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("iris.csv")
df.head()
```

```python
#here we take input data
df=df[['petal length','sepal width']]
df.head(3)
#now we sending this input inside the cluster
```

```python
model=KMeans(n_clusters=3)
y_model=model.fit_predict(df)
y_model
#here after pass centriods its finding nearby data points then make 3 group
```

```python
model.labels_
```

```python
plt.scatter(df.iloc[:,0],df.iloc[:,1],c=model.labels_,cmap='viridis')
```

```python
#now i want to check it out by groups how many raws are identified
pd.value_counts(y_model)
```

```python
#now i want to know which raw are belonging from whioch groups
df[y_model==0]
```

```python
df[y_model==1]
```

```python
df[y_model==2]
```

# Agglomeretive Hierarchical clustering
2nd algorithm of Unsupervised ML.

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
clustering= AgglomerativeClustering()
Aggclust=clustering.fit(iris.data)
Aggclust
```

```python
# here identified only two groups
Aggclust.labels_
```

```python
plt.scatter(iris.data[:,0],iris.data[:,1],c=Aggclust.labels_,cmap='viridis')
```

# Selecting No. clusters with Elbow method

```python
wcss=[]
#creat list here
k=range(1,15)
for i in k:
    km=KMeans(n_clusters=i)
    km=km.fit(df)
    wcss.append(km.inertia_)
```

```python
plt.plot(k,wcss,marker="x")
plt.xlabel('i')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```

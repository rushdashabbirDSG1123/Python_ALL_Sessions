# AUC ROC Curve
AUC (Area under the curve)
ROC(Receiver Operating characteristics)
AUC & ROC is basically Visual Representation is this also one of the matrix which is visual representation of your confusion matrix , also classification visual reprentation

ROC AUC curve represent True Posive Rate and False Positive Rate
@ sensitivity is TPR( eg. who have desiase thats are sensitivity & Recall also it is True +ve)
@1- specificity is FPR( eg. who's identified the do not have desiase thats 1-specificity it is true -ve)

#for calculating the AUC ROC curve we do have predict probability function inside the model

# Trade of between sensitivity & specificity#
#when we decrease a threshold value then we get more +ve values thus increasing the sensitivity and mean while decreasing the specificity

similarly:--- when we increase a threshold value then we get more -ve values thus increasing the specificity and mean while decreasing the sensitivity

*AUC is area under the ROC curve this score give us a good idea of how well the model performance
 so sensitivity decrease specificity increase
    specificity decrease sensitivity increase
    
#so when the sensitivity increase then will also  1- specificity increase this curve is known as ROC curve
 
#if accuracy more then more area cover in graph

so we have the 4 metrics inside the classification
1. accuracy_score
2. classification_report
3. confussion_metrics
4. AUC & ROC curve

AUC & ROC indicates how well the probabilities from the +ve classes are separated from -ve classes
AUC & ROC work with binary data classification  2 values into that predict and probability function are having it

# bais variance trade of...
a bais occurs when an algorithm has limited flaxibilityto learn from data such models pay very little attention to the training data and over simplyfly the model therefore the validation error or prediction error and training error follow similar trends such model always lead to a high error or training data and test data.
variance define the algorithm sensitivity to specific sets of data A model with a high variance pay alot of attention to training data does not generalize.  therefore

validation errort or prediction error a far apart from each other 
such model usually perform very well on training data but on test data have high error rate .. high variance causes overfitting in our model.so use regularization to fit model on our test set..

# Regularization 
regularization while traing a ML model the model can easilty overfitted or underfitted to avoid this , we use regularization in ML to properly fit a model on to out test sets 
Regularization technique help reduce the chance of overfitting & underfitting and help us to get optomal model.also reduce the error. it minimize the adjust lost function and prevent overfitting and underfitting using regularization we can fit our model on test sets and hence reduce the error.
there are 2 main types of regularization
1. Lasso regularizarion or (regressor)..(its omit & removeinternal feature & come up with problmon bais Variance trade off)
2. Ridge Regularization or (regressor)...(it minimize the gap B/w very +ve or -ve cofficeint)
3. Elastic Net Regularization or (regressor) (its combinatioj of both types)

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
#naive bayes classifier algo
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier 


from sklearn.linear_model import LinearRegression

# for logistic Regression metrics for binary data
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
# for Linear regression metrics
from sklearn.metrics import mean_squared_error, mean_absolute_error,r2_score

import warnings
warnings.filterwarnings('ignore')

#it is also linear logictic algo
from sklearn.linear_model import Lasso,Ridge,ElasticNet
```

```python
lm=LinearRegression()
lm.fit(x_train,y_train)
print(lm.score(x_train,y_train))
lmpred=lm.predict(x_test)
print(lmpred)
print("MSE:",mean_squared_error(lmpred,y_test))
print("RMSE:",np.sqrt(mean_squared_error(lmpred,y_test)))
print("r2 Score:",r2_score(lmpred,y_test))
lm.coef_
```

# aply regularization

```python
from sklearn.linear_model import Lasso,Ridge
#x_train,y_train,x_test,y_test
# will reduce the cofficient to 0 (those feature are not informative)
# alpha value could be 0.0001,0.001,0.01 value resduce all cofficients towards 0and input output
# default value of alpha =0.1
```

```python
ls=Lasso(alpha=0.0001)
ls.fit(x_train,y_train)
print(ls.score(x_train,y_train))
predlasso=ls.predict(x_test)
print("MSE:",mean_squared_error(predlasso,y_test))
print("RMSE:",np.sqrt(mean_squared_error(predlasso,y_test)))
print("r2 Score:",r2_score(predlasso,y_test))
```

```python
ls.coef_
```

# Now try to minimize cofficient variance


```python
rd=Ridge()
rd=Ridge(alpha=0.0001)
rd.fit(x_train,y_train)
print(rd.score(x_train,y_train))
predrd=rd.predict(x_test)
print("MSE:",mean_squared_error(predrd,y_test))
print("RMSE:",np.sqrt(mean_squared_error(predrd,y_test)))
print("r2 Score:",r2_score(predrd,y_test))
rd.coef_
```

# Now aply Elastic net its Combination of both

```python
from sklearn.linear_model import Lasso,Ridge,ElasticNet
```

```python
enr=ElasticNet()
enr=ElasticNet(alpha=0.0001)
enr.fit(x_train,y_train)
print(enr.score(x_train,y_train))
predenr=enr.predict(x_test)
print("MSE:",mean_squared_error(predenr,y_test))
print("RMSE:",np.sqrt(mean_squared_error(predenr,y_test)))
print("r2 Score:",r2_score(predenr,y_test))
enr.coef_
#less error in lasso in RMS in Three regularization
```
